```python
import PyPDF2

# Open the input PDF file
with open(filepath, 'rb') as file:
    # Create a PDF reader object
    reader = PyPDF2.PdfReader(file)

    # Get the total number of pages
    num_pages = len(reader.pages)

    # Calculate the number of output files needed
    num_output_files = (num_pages + 99) // 100

    # Loop through the output files
    for i in range(num_output_files):
        # Create a PDF writer object
        writer = PyPDF2.PdfWriter()

        # Determine the start and end pages for this output file
        start_page = i * 100
        end_page = min((i + 1) * 100, num_pages)

        # Add the pages to the output file
        for j in range(start_page, end_page):
            page = reader.pages[j] 
            writer.add_page(page)

        # Write the output file
        output_filename = f'output_{i+1}.pdf'
        with open(output_filename, 'wb') as output_file:
            writer.write(output_file)
            print(f'Created {output_filename} with {end_page - start_page} pages')
```

```python
import re

def keyword_search(search_text, parsed_text):
    results = []
    
    for key, text in parsed_text.items():
        print("Serching part", key)
        pattern = r'\b' + re.escape(search_text) + r'\b'
        matches = re.finditer(pattern, text, re.IGNORECASE)
        for match in matches:
            results.append((key, match.start(), match.end()))
    
    return results

def print_keyword_matches(search_text, parsed_text):
    keyword_results = keyword_search(search_text, parsed_text)
    
    print(f"Keyword search results for '{search_text}':")
    match_count = 10
    print(len(keyword_results))

    for result in keyword_results:
        key, start, end = result
        text = parsed_text[key]
        context_start = max(0, start - 50)
        context_end = min(len(text), end + 1000)
        match_text = text[start:end]
        context = text[context_start:context_end]
        print(f"\n#Match {match_count}")
        print(f"{context[:context.index(match_text)]}**{match_text}**{context[context.index(match_text) + len(match_text):]}")
        match_count += 1

# Example usage

search_text = "residence in india"

print_keyword_matches(search_text, parsed_text)
```
